```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.myntra.com/men-casual-shirts")
```

```python
links=[]
```

```python
for i in range(8):
    time.sleep(2)
    links_try=driver.find_elements(By.XPATH,"""//*[@class="product-base"]/a""")
    time.sleep(5)
    for i in links_try:
     links.append(i.get_attribute("href"))
    try:
      button=driver.find_element(By.XPATH,"""//*[@class="pagination-next"]""")
    except:
        time.sleep(3)
        button=driver.find_element(By.XPATH,"""//*[@class="pagination-next"]""")
    button.click()
```

```python
for i in links_try:
    links.append(i.get_attribute("href"))
```

```python
len(links)
```

```python
name=[]
stars=[]
no_of_ratings=[]
price=[]
discount=[]
sleeve=[]
fit=[]
collar=[]
```

```python
for i in links:
    driver.get(i)
    time.sleep(4)
    try:
        name.append(driver.find_element(By.XPATH,"""//*[@class="pdp-name"]""").text)
    except:
        name.append(None)
    try:
        stars.append(driver.find_element(By.XPATH,"""//*[@class="index-overallRating"]/div""").text)
    except:
        stars.append(None)
    try:
        no_of_ratings.append(driver.find_element(By.XPATH,"""//*[@class="index-ratingsCount"]""").text)
    except:
        no_of_ratings.append(None)
    try:
        price.append(driver.find_element(By.XPATH,"""//*[@class="pdp-price"]""").text)
    except:
        price.append(None)
    try:
        discount.append(driver.find_element(By.XPATH,"""//*[@class="pdp-discount"]""").text)
    except:
        discount.append(None)
    try:
        sleeve.append(driver.find_element(By.XPATH,"""//*[@class="index-row"]/div[2]""").text)
    except:
        sleeve.append(None)
    try:
        fit.append(driver.find_element(By.XPATH,"""//*[@class="index-row"][3]/div[2]""").text)
    except:
        fit.append(None)
    try:
        collar.append(driver.find_element(By.XPATH,"""//*[@class="index-row"][2]/div[2]""").text)
    except:
        collar.append(None)
    time.sleep(3)
    print("success")
```
