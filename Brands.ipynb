```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
driver=webdriver.Chrome()
driver.get("https://www.myntra.com/men-casual-shirts")
```

```python
links=[]
brands=[]
```

```python
for i in range(8):
    time.sleep(2)
    links_try=driver.find_elements(By.XPATH,"""//*[@class="product-base"]/a""")
    brands_try= driver.find_elements(By.XPATH,"""//*[@class="product-brand"]""")
    time.sleep(10)
    for i in links_try:
     links.append(i.get_attribute("href"))
     time.sleep(1)   
    time.sleep(10)    
    for i in brands_try:
     brands.append(i.text) 
     time.sleep(1)   
    time.sleep(10) 
    try:
      button=driver.find_element(By.XPATH,"""//*[@class="pagination-next"]""")
    except:
        time.sleep(3)
        button=driver.find_element(By.XPATH,"""//*[@class="pagination-next"]""")
    button.click()
```
